Chaque noeud se repère par son indice *n* dans la liste, son fils gauche se trouvant alors à l'indice *2n + 1* et son fils droit à l'indice *2n + 2*.

```python
def creation_arbre(r,profondeur):
    """r : la racine (str ou int). la profondeur de l’arbre (int)"""
    Arbre = [r]+[None for i in range(2**(profondeur+1)-2)]
    return Arbre

def insertion_noeud(arbre,n,fg,fd):
    """Insére les noeuds et leurs enfants dans l’arbre"""
    indice = arbre.index(n)
    arbre[2*indice+1] = fg
    arbre[2*indice+2] = fd

# création de l’arbre
arbre = creation_arbre("r",5)
# ajout des noeuds par niveau de gauche à droite
insertion_noeud(arbre,"r","a","b")
insertion_noeud(arbre,"a","c","d")
insertion_noeud(arbre,"b","e","f")
insertion_noeud(arbre,"c",None,"h")
insertion_noeud(arbre,"d","i","j")
insertion_noeud(arbre,"e","k",None)
insertion_noeud(arbre,"f",None,None)
insertion_noeud(arbre,"h",None,None)
insertion_noeud(arbre,"i",None,None)
insertion_noeud(arbre,"j","m",None)
insertion_noeud(arbre,"k",None,None)
insertion_noeud(arbre,"m",None,None)

#pour vérifier
print(len(arbre))
print(arbre)
```

```python
def file_vide():
    return []

def est_vide(F):
    return F == []

def enfile(F, element):
    F.append(element)

def defile(F):
    return F.pop(0)

def first(F):
    return F[0]
```

```python
a = ["a", ["b", ["d", [], []], ["e", [], []]], ["c", ["f", [], []], ["g", [], []]]]
b = ["a", "b", "c", "d", "e", "f", "g", None, None, None, None, None, None, None, None]

def parcoursLargeur_a(arbre):
    f = file_vide()
    liste = []
    enfile(f, arbre)
    while not est_vide(f):
        tmp = defile(f)
        liste.append(tmp[0])
        if tmp[1] != []:
            enfile(f,tmp[1])
        if tmp[2] != []:
            enfile(f, tmp[2])
    return liste

print(parcoursLargeur_a(a))

def parcoursLargeur_b(arbre):
    f = file_vide()
    liste = []
    for i in arbre:
        enfile(f, i)
    while not est_vide(f):
        tmp = defile(f)
        if not tmp is None:
            liste.append(tmp)
    return liste

print(parcoursLargeur_b(b))
```

```python
class Noeud():
    def __init__(self, r, fg = None, fd = None):
        self.racine = r
        self.fgauche = fg
        self.fdroit = fd
    
    def estFeuille(self):
        return self.fgauche is None and self.fdroit is None
    
    def hauteur(self):
        fg = self.fgauche
        hg = 1
        while not fg is None:
            fg = fg.fgauche
            hg += 1
        fd = self.fdroit
        hd = 1
        while not fd is None:
            fd = fd.fdroit
            hd += 1
        if hg >= hd:
            return hg
        else:
            return hd
        
    def parcoursPrefixe(self):
        print("'{}'".format(self.racine), end = " ")
        if not self.fgauche is None:
            self.fgauche.parcoursPrefixe()
        if not self.fdroit is None:
            self.fdroit.parcoursPrefixe()
            
    def parcoursInfixe(self):
        if not self.fgauche is None:
            self.fgauche.parcoursPrefixe()
        print("'{}'".format(self.racine), end = " ")
        if not self.fdroit is None:
            self.fdroit.parcoursPrefixe()
    
    def parcoursSuffixe(self):
        if not self.fgauche is None:
            self.fgauche.parcoursPrefixe()
        if not self.fdroit is None:
            self.fdroit.parcoursPrefixe()
        print("'{}'".format(self.racine), end = " ")

        
def hauteur(arbre):
    return arbre.hauteur()
        
arbre = Noeud("a", fg = Noeud("b", fg = Noeud("d"), fd = Noeud("e")), fd = Noeud("c", fg = Noeud("f"), fd = Noeud("g")))
print(arbre.hauteur())
print(arbre.estFeuille())
print("\n")
arbre.parcoursPrefixe()
print("\n")
arbre.parcoursInfixe()
print("\n")
arbre.parcoursSuffixe()
```
