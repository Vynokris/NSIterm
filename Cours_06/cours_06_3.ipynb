## IV. Les arbres binaires de recherches

Un arbre binaire de recherche (ABR) est une structure de donnée composée de noeuds. Chaque noeud a au plus 2 enfants ordonnés d'une manière particulière:
* les enfants,à gauche d'un noeud ont des valeurs inférieures à lui
* les enfants à droite d'un noeud ont des valeurs supérieures à lui

Et cela doit être vrai pour chaque noeud de l'arbre.

Nous travaillerons avec la classe Noeud ci-dessous dans laquelle une variable d'instance parent a été rajoutée:

```python
class Noeud:
    # Le constructeur
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right
        self.parent = None
        
    def __str__(self):
        return str(self.value)
    
    def estFeuille(self):
        if not self.left and not self.right:
            return True
        else:
            return False
    
    def insert(self, valeur):
        if valeur < self.value:
            if self.left is None:
                self.left = Noeud(valeur)
                self.left.parent = self
            else:
                self.left.insert(valeur)
        elif valeur > self.value:
            if self.right is None:
                self.right = Noeud(valeur)
                self.right.parent =self
            else:
                self.right.insert(valeur)
    
    def parcoursPrefixe(self, output=[]):
        output.append(self.value)
        #print("'{}'".format(self.value), end = " ")
        if not self.left is None:
            self.left.parcoursPrefixe(output = output)
        if not self.right is None:
            self.right.parcoursPrefixe(output = output)
        return output
            
    def parcoursInfixe(self, output=[]):
        if not self.left is None:
            self.left.parcoursInfixe(output = output)
        output.append(self.value)
        #print("'{}'".format(self.value), end = " ")
        if not self.right is None:
            self.right.parcoursInfixe(output = output)
        return output
    
    def parcoursSuffixe(self, output=[]):
        if not self.left is None:
            self.left.parcoursSuffixe(output = output)
        if not self.right is None:
            self.right.parcoursSuffixe(output = output)
        #print("'{}'".format(self.value), end = " ")
        output.append(self.value)
        return output
        
```

```python
def insert(self, valeur):
    if valeur < self.value:
        if self.left is None:
            self.left = Noeud(valeur)
            self.left.parent = self
        else:
            self.left.insert(valeur)
    elif valeur > self.value:
        if self.right is None:
            self.right = Noeud(valeur)
            self.right.parent =self
        else:
            self.right.insert(valeur)
```

4. Ecrire un programme qui utilise une structure d'ABR pour trier une liste d'entier

```python
l = [1, 3, 7, 8, 5, 4, 2, 9, 0, 6]

def tree_from_list(liste):
    arbre = Noeud(liste[0])
    for i in range(1, len(liste)):
        arbre.insert(liste[i])
    return arbre

def triABR(liste):
    arbre = tree_from_list(liste)
    return arbre.parcoursInfixe(output = [])

print(triABR(l))
```

5. Ecrire un programme qui utilise une structure d'ABR pour trouver le maximum d'une liste d'entier

```python
def maxABR(liste):
    arbre = tree_from_list(liste)
    max = arbre
    while not max.right is None:
        max = max.right
    return max.value

print(maxABR(l))
```

6. Ecrire un programme qui utilise une structure d'ABR pour trouver le minimum d'une liste d'entiers

```python
def minABR(liste):
    arbre = tree_from_list(liste)
    min = arbre
    while not min.left is None:
        min = min.left
    return min.value

print(minABR(l))
```

7. Ecrire un programme qui utilise une structure d'ABR pour rechercher une valeur dans la liste:
L = ['chat', 'chien', 'souris', 'araignée', 'crapaud', 'grenouille', 'lézard', 'zèbre']

```python
def findABR(liste, valeur):
    arbre = tree_from_list(liste)
    noeud = arbre
    while not noeud.value == valeur:
        if noeud.left is None and noeud.right is None:
            return False
        elif valeur > noeud.value and noeud.right is None:
            return False
        elif valeur < noeud.value and noeud.left is None:
            return False
        
        if valeur > noeud.value:
            noeud = noeud.right
        elif valeur < noeud.value:
            noeud = noeud.left
    return True

L = ['chat', 'chien', 'souris', 'araignée', 'crapaud', 'grenouille', 'lézard', 'zèbre']
print(findABR(L, "grenouille"))
```
